```python
from tqdm import tqdm
import os
from glob import glob
import re
import numpy as np
import pandas as pd

from my_module import *
```

```python
def add_critic_scores(df):
    df_score = pd.DataFrame(df['Name'])
    df_review = pd.DataFrame(df['Name'])
    df_score['Status'] = np.nan
    df_review['Status'] = np.nan
    add_link = '/critic-reviews'
    for counter, root_url in tqdm(list(enumerate(df['Link']))):
        try: driver = open_url(root_url+add_link)
        except: 
            df_score['Status'].iloc[counter] = 'WebDriverException'
            df_review['Status'].iloc[counter] = 'WebDriverException'
            continue
        reviws_els = driver.find_elements(by=By.CLASS_NAME, value='critic_review')
        for i in range(len(reviws_els)):
            text = reviws_els[i].text.split('\n')
            if text[2].isnumeric():
                score = int(text[2])
                review = text[3]
                critic = text[0]
            elif text[1].isnumeric():
                score = int(text[1])
                review = text[2]
                critic = text[0]
            else: 
                score = '...'
                review = text[2]
                critic = text[0]
            try: 
                df_score[critic].iloc[counter] = score
                df_review[critic].iloc[counter] = review
            except: 
                df_score[critic] = np.nan
                df_score[critic].iloc[counter] = score
                df_review[critic] = np.nan
                df_review[critic].iloc[counter] = review
                
        driver.close()
    return df_score, df_review
```

```python
glob('dataframes/*.csv')
```

```python
CSVs = glob('dataframes/*.csv')
for i,path in enumerate(CSVs):
    platform_df = pd.read_csv(path).drop('Unnamed: 0', axis=1)
    df_score, df_review = add_critic_scores(platform_df)
    st = CSVs[i].split('_')[0].split('\\')[-1]
    df_score.to_csv(f'dataframes/{st}_scores.csv')
    df_review.to_csv(f'dataframes/{st}_reviews.csv')
```
