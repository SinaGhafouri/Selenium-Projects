```python
root_link_xsx = 'https://www.metacritic.com/browse/games/score/metascore/all/xbox-series-x?sort=desc'
root_link_ps5 = 'https://www.metacritic.com/browse/games/score/metascore/all/ps5?sort=desc'
root_link_switch = 'https://www.metacritic.com/browse/games/score/metascore/all/switch?sort=desc'
root_link_xb1 = 'https://www.metacritic.com/browse/games/score/metascore/all/xboxone?sort=desc'
root_link_pc = 'https://www.metacritic.com/browse/games/score/metascore/all/pc?sort=desc'
root_link_ps4 = 'https://www.metacritic.com/browse/games/score/metascore/all/ps4?sort=desc'
```

```python
def get_links(driver):
    links = []
    for i in range(1,10):
        try:
            el = driver.find_element(by=By.XPATH, value=f'//*[@id="main_content"]/div[1]/div[2]/div/div[1]/div/div[2]/table/tbody/tr[{str(i)}]/td[2]/a')
            # print(i, el.text, el.get_attribute('href'))
            links.append(el.get_attribute('href'))
        except: pass
    for i in range(1,10):
        try:
            el = driver.find_element(by=By.XPATH, value=f'//*[@id="main_content"]/div[1]/div[2]/div/div[1]/div/div[4]/table/tbody/tr[{str(i)}]/td[2]/a')
            # print(i, el.text, el.get_attribute('href'))
            links.append(el.get_attribute('href'))
        except: pass    
    for i in range(1,10):
        try:
            el = driver.find_element(by=By.XPATH, value=f'//*[@id="main_content"]/div[1]/div[2]/div/div[1]/div/div[6]/table/tbody/tr[{str(i)}]/td[2]/a')
            # print(i, el.text, el.get_attribute('href'))
            links.append(el.get_attribute('href'))
        except: pass
    for i in range(1,200):
        try:
            el = driver.find_element(by=By.XPATH, value=f'//*[@id="main_content"]/div[1]/div[2]/div/div[1]/div/div[8]/table/tbody/tr[{str(i)}]/td[2]/a')
            # print(i, el.text, el.get_attribute('href'))
            links.append(el.get_attribute('href'))
        except: pass
    return links

# 'https://www.metacritic.com/browse/games/score/metascore/all/pc?sort=desc'
# 'https://www.metacritic.com/browse/games/score/metascore/all/pc?sort=desc&page=52'
def get_all_links(root_url):
    global pages_els
    driver = open_url(root_url)
    links = get_links(driver)
    all_pages_links = []
    all_pages_links.extend(links)
    pages_els = driver.find_elements(by=By.CLASS_NAME, value="page_num")
    last_page = int(pages_els[-1].text)
    # next_pages_links = [pages_els[i+1].get_attribute('href') for i in range(len(pages_els)-1)]
    driver.close()
    for next_page in range(1,last_page):
        next_page_url = root_url+f'&page={next_page}'
        driver = open_url(next_page_url)
        links = get_links(driver)
        all_pages_links.extend(links)
        driver.close()
    return all_pages_links
```
